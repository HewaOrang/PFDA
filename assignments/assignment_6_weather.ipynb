```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv(("https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv"), skiprows=23)

print(df.head(20))
```

```python
df.info()
```

```python
# Convert date column to datetime
# https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
df['date'] = pd.to_datetime(df['date'])

# Set date as index 
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html

df = df.set_index('date') 

print(df.head())
```

```python
#  Temperature Over Time
plt.figure()
df['temp'].plot()
plt.title('Hourly Temperature')
plt.ylabel('Temperature (C)')
plt.xlabel('Date')
plt.show()
```

```python
# Daily Mean Temperature Over Time
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/09_timeseries.html

daily_mean_temp = df['temp'].resample('D').mean() # resample by day

plt.figure()
daily_mean_temp.plot()
plt.title('Daily Mean Temperature')
plt.ylabel('Temperature (C)')
plt.xlabel('Date')
plt.show()
```

```python
# Monthly Mean Temperature Over Time
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/09_timeseries.html

monthly_mean_temp = df['temp'].resample('ME').mean() # resample by month end 

plt.figure()
monthly_mean_temp.plot()
plt.title('Monthly Mean Temperature')
plt.ylabel('Temperature (C)')
plt.xlabel('Date')
plt.show()
```

```python
# Convert 'wdsp' to numeric, coercing errors to NaN
# https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html

df['wdsp'] = pd.to_numeric(df['wdsp'], errors='coerce') # Convert wind speed to numeric, coercing errors to NaN
```

```python
# wind speed over time
plt.figure()
df['wdsp'].plot()
plt.title('Hourly Wind Speed')
plt.ylabel('Wind Speed (knots)')
plt.xlabel('Date')
plt.show()
```

```python
# 24-Hour Rolling Mean Wind Speed
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html#pandas.DataFrame.rolling

rolling_24h_wind = df['wdsp'].rolling(window=24, min_periods=1).mean()      # Since the data is hourly, a 24-sample window represents 24 hours
                                                                            # min_periods=1 ensures early values are still calculated

plt.figure()
rolling_24h_wind.plot()
plt.title('24-Hour Rolling Mean Wind Speed')
plt.ylabel('Wind Speed (knots)')
plt.xlabel('Date')
plt.show()
```

```python
# Daily Maximum Wind Speed

daily_max_wind = df['wdsp'].resample('D').max() # resample by day

plt.figure()
daily_max_wind.plot()
plt.title('Daily Maximum Wind Speed')
plt.ylabel('Wind Speed (knots)')
plt.xlabel('Date')
plt.show()
```

```python
# Monthly Mean of Daily Maximum Wind Speeds
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/09_timeseries.html

monthly_mean_daily_max_wind = daily_max_wind.resample('ME').mean() # resample by month

plt.figure()
monthly_mean_daily_max_wind.plot()
plt.title('Monthly Mean of Daily Maximum Wind Speeds')
plt.ylabel('Wind Speed (knots)')
plt.xlabel('Date')
plt.show()
```
