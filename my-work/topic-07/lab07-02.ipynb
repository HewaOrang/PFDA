```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# load the dataset
dataset = sns.load_dataset('tips')

# the for debugging
#print(dataset.head())
sns.set_style('whitegrid')
sns.lmplot(x= 'total_bill', y='tip', order=3, data=dataset)
plt.show()
```

```python
sns.lmplot(x="size", y="tip", data=dataset)
```

```python
sns.lmplot(x="size", y="tip", data=dataset, x_jitter=0.5)
```

```python
sns.lmplot(x="size", y="tip", data=dataset,  x_estimator=np.mean)
```
